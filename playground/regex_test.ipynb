```python
import re
```

```python
test_paragraph = (
    "Hello-World!\r\n"           # 1. Hyphenated word + Windows Linebreak
    "This is a test.\n"          # 2. Standard spaces + Unix Linebreak
    "Sector7 is invalid.\t"      # 3. Alphanumeric (no boundary) + Tab
    "R2D2 can't fly.\r"          # 4. Mixed digits + apostrophe + Old Mac Linebreak
    "End_of_line"                # 5. Underscores (considered word chars)
)
```

```python
re.findall(r"\b[a-zA-Z]+[\d]*(?:[-'][a-zA-Z]+[\d]*)*\b", string=test_paragraph)
```
